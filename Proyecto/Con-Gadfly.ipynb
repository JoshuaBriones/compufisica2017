```julia
using Gadfly
```

```julia
a=0.25 #m
q=1.9e-12 #C
k=9.0e9 #Nm²/C²
p=2*a*q #mC
k*p
```

```julia
plot(z=(x,y) -> (k*p)*((1/(((y+a)^2+x^2+1.0e-10)^(1/2)))-(1/(((y-a)^2+x^2+1.0e-10)^(1/2)))),
     x=collect(-0.5:1/4:0.5), y=collect(-0.5:1/4:0.5), Geom.contour(levels=16))
```

```julia
coord = Coord.cartesian(xmin=-0.5, xmax=0.5, ymin=-0.5, ymax=0.5)
plot(coord, z=(x,y)->(k*p)*((1/(((y+a)^2+x^2+1.0e-10)^(1/2)))-(1/(((y-a)^2+x^2+1.0e-10)^(1/2)))),
        xmin=[-0.5], xmax=[0.5], ymin=[-0.5], ymax=[0.5],
        Geom.vectorfield(scale=0.1,samples=16), Geom.contour(levels=15),
        Scale.x_continuous(minvalue=-0.5, maxvalue=0.5),
        Scale.y_continuous(minvalue=-0.5, maxvalue=0.5),
        Guide.xlabel("x"), Guide.ylabel("y"), Guide.colorkey("z")
    )
```
